## Summary

## Get encode seq for visualize

```python
data = sio.loadmat("../results/feature_donor_seq.mat")
Y = scipy.matrix(data['Y'])
encodeSeq1 = data['encodeSeq1']
encodeSeq2 = data['encodeSeq2']
```

## Plot the sequences to find the pattern

```python
print('All the sequences')
viz_sequence.plot_weights(encodeSeq1.mean(0))
#plt.title('random')
print('enriched for SD1 - i=0')
viz_sequence.plot_weights(encodeSeq1[np.where(Y[:,0] > 0.9)[0]].mean(0))
print('enriched for SD1 - i=0 - 2nd sequence')
viz_sequence.plot_weights(encodeSeq2[np.where(Y[:,0] > 0.9)[0]].mean(0))
print('depleted for SD1 - i=0')
viz_sequence.plot_weights(encodeSeq1[np.where(Y[:,0] < 0.05)[0]].mean(0))
print('depleted for SD1 - i=0 - seq2')
viz_sequence.plot_weights(encodeSeq2[np.where(Y[:,0] < 0.05)[0]].mean(0))

print('enriched for SD2 - i=44 - 1st sequence')
viz_sequence.plot_weights(encodeSeq1[np.where(Y[:,44] > 0.9)[0]].mean(0))
print('enriched for SD2 - i=44 - 2nd sequence')
viz_sequence.plot_weights(encodeSeq2[np.where(Y[:,44] > 0.9)[0]].mean(0))
print('depleted for SD2 - i=44 - seq1')
viz_sequence.plot_weights(encodeSeq1[np.where(Y[:,44] < 0.05)[0]].mean(0))
print('depleted for SD2 - i=44')
viz_sequence.plot_weights(encodeSeq2[np.where(Y[:,44] < 0.05)[0]].mean(0))

print('enriched for SDnew - i=9')
viz_sequence.plot_weights(encodeSeq1[np.where(Y[:,9] > 0.9)[0]].mean(0))
print('depleted for SDnew - i=9')
viz_sequence.plot_weights(encodeSeq1[np.where(Y[:,9] < 0.05)[0]].mean(0))

print('enriched for SDnew - i=18')
viz_sequence.plot_weights(encodeSeq1[np.where(Y[:,18] > 0.9)[0]].mean(0))

print('enriched for SDnew - i=50')
viz_sequence.plot_weights(encodeSeq2[np.where(Y[:,50] > 0.9)[0]].mean(0))

print('enriched for SDnew - i=58')
viz_sequence.plot_weights(encodeSeq2[np.where(Y[:,58] > 0.9)[0]].mean(0))


print('enriched for SDcrypt - i=79')
viz_sequence.plot_weights(encodeSeq2[np.where(Y[:,79] > 0.9)[0]].mean(0))
```
