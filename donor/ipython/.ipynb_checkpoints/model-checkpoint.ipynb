```python
train, test = cm.get_data(os.path.abspath("../results/feature_donor_seq.mat"),
                          os.path.abspath('../results/paritions.npy'))
```

```python
m = cm.get_model(train,             
              filters=128, # use powers of 2 - (32, 64, 128)
              motif_width=15, # use odd numbers - 11 or 15
              use_1by1_conv=True, # hp.choice(True, False(           
              # dense layers
              task_specific={'l1_1': 2.3794563161482e-12, 
                      'l1_2' : 1.3638773409491815e-12,
                      'l1_3' :6.156281322944606e-07, 
                      },lr=0.00010144196294995054
)

# train the model
m.fit(train[0], y=train[1], batch_size=32,
          epochs=300,

          validation_split=.2, 
          callbacks=[EarlyStopping(patience=10)])

y_pred = m.predict(test[0])

evaluate_r2_score(y_pred, test[1])
```
